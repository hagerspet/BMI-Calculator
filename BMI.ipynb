```python
name=input("Enter your name:")
w=float(input("Enter Your weight (KG)"))
h=float(input("Enter Your height (Meter)"))
bmi=w/h**2
bmi
if bmi>0 and bmi<=16:
  print("Severe Thinness",name,"You should eating well")
elif bmi>16 and bmi<=17:
  print("Moderate Thinness")
elif bmi>17 and bmi<=18.5:
  print("Mild Thinness")
elif bmi>18.5 and bmi<=25:
  print("Normal")
elif bmi>25 and bmi<=30:
  print("Overweight",name,"You should do more execries")
elif bmi>30:
  print("Obese")
else:
  print("Enter valid inp  ut")
```
